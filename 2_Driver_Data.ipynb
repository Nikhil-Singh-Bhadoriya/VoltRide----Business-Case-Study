```python
# Group by City and sum the Total_Rides_Last_30_Days
rides_by_city = df1.groupby('City')['Total_Rides_Last_30_Days'].sum().sort_values(ascending=False)
print(rides_by_city)

# Visualize with a bar plot
plt.figure(figsize=(10, 6))
rides_by_city.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Total Rides Last 30 Days by City')
plt.xlabel('City')
plt.ylabel('Total Rides')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Univariate Analysis for all columns in df1

# Select only numeric columns for univariate analysis
numeric_cols = df1.select_dtypes(include=[np.number]).columns.tolist()

# Create figure with multiple subplots for distribution analysis
fig, axes = plt.subplots(len(numeric_cols), 3, figsize=(18, 5*len(numeric_cols)))

for idx, col in enumerate(numeric_cols):
    # Histogram
    axes[idx, 0].hist(df1[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    axes[idx, 0].set_title(f'Histogram: {col}')
    axes[idx, 0].set_xlabel(col)
    axes[idx, 0].set_ylabel('Frequency')
    axes[idx, 0].grid(True, alpha=0.3)
    
    # Box plot
    axes[idx, 1].boxplot(df1[col].dropna(), vert=True)
    axes[idx, 1].set_title(f'Box Plot: {col}')
    axes[idx, 1].set_ylabel(col)
    axes[idx, 1].grid(True, alpha=0.3)
    
    # Density plot (KDE)
    df1[col].dropna().plot(kind='density', ax=axes[idx, 2], color='green')
    axes[idx, 2].set_title(f'Density Plot: {col}')
    axes[idx, 2].set_xlabel(col)
    axes[idx, 2].set_ylabel('Density')
    axes[idx, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Statistical summary for numeric columns
print("\n=== Statistical Summary ===")
print(df1[numeric_cols].describe())

# Categorical columns analysis
categorical_cols = df1.select_dtypes(include=['object']).columns.tolist()

if categorical_cols:
    print("\n=== Categorical Columns Analysis ===")
    
    for col in categorical_cols:
        print(f"\n{col} - Value Counts:")
        print(df1[col].value_counts())
        
        # Bar plot for categorical variables
        plt.figure(figsize=(10, 6))
        df1[col].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
        plt.title(f'Frequency Distribution: {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.grid(True, alpha=0.3, axis='y')
        plt.tight_layout()
        plt.show()

# Check for missing values
print("\n=== Missing Values ===")
print(df1.isnull().sum())

# Skewness and Kurtosis
print("\n=== Skewness and Kurtosis ===")
for col in numeric_cols:
    skewness = df1[col].skew()
    kurtosis = df1[col].kurtosis()
    print(f"{col}: Skewness = {skewness:.2f}, Kurtosis = {kurtosis:.2f}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Trivariate Analysis for all variables in the dataset

# 1. 3D Scatter plots for selected numeric variable combinations

fig = plt.figure(figsize=(20, 12))

# Select combinations of three numeric variables
trivariate_combinations = [
    ('Experience_Months', 'Avg_Rating', 'Cancellation_Rate_%'),
    ('Total_Rides_Last_30_Days', 'Experience_Months', 'Avg_Rating'),
    ('Avg_Battery_Consumption_km', 'Total_Rides_Last_30_Days', 'Cancellation_Rate_%'),
    ('Experience_Months', 'Total_Rides_Last_30_Days', 'Cancellation_Rate_%'),
]

for idx, (var1, var2, var3) in enumerate(trivariate_combinations, 1):
    ax = fig.add_subplot(2, 2, idx, projection='3d')
    
    scatter = ax.scatter(df1[var1], df1[var2], df1[var3], 
                        c=df1[var3], cmap='viridis', s=50, alpha=0.6, edgecolors='k')
    
    ax.set_xlabel(var1, fontsize=10)
    ax.set_ylabel(var2, fontsize=10)
    ax.set_zlabel(var3, fontsize=10)
    ax.set_title(f'3D: {var1} vs {var2} vs {var3}', fontsize=11)
    plt.colorbar(scatter, ax=ax, label=var3)

plt.tight_layout()
plt.show()

# 2. Trivariate analysis: Numeric x Numeric x Categorical
print("\n=== Trivariate Analysis: Numeric x Numeric x Categorical ===\n")

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.ravel()

trivariate_numeric_cat = [
    ('Experience_Months', 'Total_Rides_Last_30_Days', 'Cancellation_Rate_%', 'City'),
    ('Avg_Rating', 'Experience_Months', 'Cancellation_Rate_%', 'City'),
    ('Avg_Battery_Consumption_km', 'Total_Rides_Last_30_Days', 'Avg_Rating', 'City'),
    ('Experience_Months', 'Avg_Rating', 'Total_Rides_Last_30_Days', 'City'),
]

for idx, (var1, var2, var3, cat_var) in enumerate(trivariate_numeric_cat):
    cities = df1[cat_var].unique()
    colors = ['red', 'blue', 'green', 'orange']
    
    for city, color in zip(cities, colors):
        city_mask = df1[cat_var] == city
        axes[idx].scatter(df1[city_mask][var1], df1[city_mask][var2], 
                         s=df1[city_mask][var3]*2, alpha=0.5, 
                         label=city, color=color, edgecolors='k')
    
    axes[idx].set_xlabel(var1, fontsize=10)
    axes[idx].set_ylabel(var2, fontsize=10)
    axes[idx].set_title(f'{var1} vs {var2} (bubble size: {var3}) by {cat_var}', fontsize=10)
    axes[idx].legend()
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3. Trivariate analysis with heatmaps by category
print("\n=== Trivariate Heatmap Analysis by City ===\n")

for city in df1['City'].unique():
    city_data = df1[df1['City'] == city]
    
    fig, ax = plt.subplots(figsize=(10, 8))
    corr_matrix = city_data[numeric_cols_for_analysis].corr()
    
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0,
                square=True, linewidths=1, ax=ax, cbar_kws={"shrink": 0.8})
    ax.set_title(f'Correlation Matrix for {city}', fontsize=12)
    plt.tight_layout()
    plt.show()

# 4. Interaction effects: Numeric x Numeric x Numeric with binning
print("\n=== Trivariate Interaction Analysis ===\n")

var1, var2, var3 = 'Experience_Months', 'Avg_Rating', 'Cancellation_Rate_%'

# Create bins for better visualization
df1['exp_bin'] = pd.cut(df1[var1], bins=3, labels=['Low', 'Medium', 'High'])
df1['rating_bin'] = pd.cut(df1[var2], bins=3, labels=['Low', 'Medium', 'High'])

fig, ax = plt.subplots(figsize=(12, 8))

interaction_data = df1.groupby(['exp_bin', 'rating_bin'])[var3].mean().unstack()
sns.heatmap(interaction_data, annot=True, fmt='.2f', cmap='RdYlGn_r', ax=ax,
           cbar_kws={"shrink": 0.8})
ax.set_title(f'Interaction Effect: {var1} x {var2} on {var3}', fontsize=12)
ax.set_xlabel(f'{var2} Bins', fontsize=10)
ax.set_ylabel(f'{var1} Bins', fontsize=10)
plt.tight_layout()
plt.show()

# 5. Categorical x Categorical x Numeric analysis
print("\n=== Categorical x Categorical Interaction Analysis ===\n")

# Create experience level bins
df1['experience_level'] = pd.cut(df1['Experience_Months'], 
                                 bins=[0, 20, 40, 100], 
                                 labels=['Junior', 'Mid', 'Senior'])

for num_var in ['Avg_Rating', 'Cancellation_Rate_%', 'Total_Rides_Last_30_Days']:
    fig, ax = plt.subplots(figsize=(12, 6))
    
    pivot_data = df1.pivot_table(values=num_var, 
                                 index='experience_level', 
                                 columns='City', 
                                 aggfunc='mean')
    
    pivot_data.plot(kind='bar', ax=ax, edgecolor='black', width=0.8)
    ax.set_title(f'{num_var} by Experience Level and City', fontsize=12)
    ax.set_xlabel('Experience Level', fontsize=10)
    ax.set_ylabel(f'Mean {num_var}', fontsize=10)
    ax.legend(title='City', fontsize=9)
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.show()

# 6. Statistical summary for trivariate relationships
print("\n=== Trivariate Statistical Analysis ===\n")

for city in df1['City'].unique():
    city_data = df1[df1['City'] == city]
    print(f"\n{'='*50}")
    print(f"City: {city}")
    print(f"{'='*50}")
    print(f"Sample size: {len(city_data)}")
    print(f"\nMean values:")
    print(city_data[numeric_cols_for_analysis].mean())
    print(f"\nMedian Cancellation Rate: {city_data['Cancellation_Rate_%'].median():.2f}%")
    print(f"Mean Avg Rating: {city_data['Avg_Rating'].mean():.2f}")
    print(f"Mean Total Rides: {city_data['Total_Rides_Last_30_Days'].mean():.2f}")

print("\n=== Trivariate Analysis Complete ===")
```
